```python
#the data obtained from the survey conducted by myself
import pandas as pd
df_forms = pd.read_csv("form_responses - Form Responses 1.csv")
df_forms.head()
```

```python
df_forms.columns = [
    "timestamp", "age", "gender", "sleep_hours", "bedtime_consistency",
    "bedtime", "wake_time", "stress_level", "anxiety_level",
    "efficiency_level", "sleep_difficulty"
]
df_forms = df_forms.drop(columns=["timestamp"])
df_forms.head()
```

```python
def convert_time(t):
    try:
        return int(t)
    except:
        return None

df_forms["bedtime_hour"] = df_forms["bedtime"].apply(convert_time)
df_forms["wake_hour"] = df_forms["wake_time"].apply(convert_time)
def bucket_sleep(x):
    x = str(x).lower()
    if "less" in x or "4" in x:
        return "Very Short"
    elif "4-6" in x:
        return "Short"
    elif "6-8" in x:
        return "Normal"
    elif "8" in x:
        return "Long"
    else:
        return "Other"

df_forms["sleep_bucket"] = df_forms["sleep_hours"].apply(bucket_sleep)
for col in ["stress_level", "anxiety_level", "efficiency_level"]:
    df_forms[col + "_norm"] = df_forms[col] / 10.0
df_forms["sleep_difficulty_score"] = df_forms["sleep_difficulty"].map({
    "Yes": 1,
    "Sometimes": 0.5,
    "No": 0
})
df_forms.head()
df_forms.info()
```

```python
df_public.columns = df_public.columns.str.strip().str.lower().str.replace(" ", "_")
columns_to_keep = [
    "subject", "day", "step_count", "distance_travelled_(km)",
    "actual_sleep_duration_(minutes)", "sleep_onset_latency_(minutes)",
    "duration_in_bed_(minutes)", "efficiency_(%)",
    "deep_sleep_duration_(minutes)_(minutes)", "rem_duration_(minutes)",
    "awake_percentage_(%)", "light_sleep_percentage_(%)"
]
df_public = df_public[columns_to_keep]

df_public = df_public.rename(columns={
    "actual_sleep_duration_(minutes)": "sleep_duration",
    "duration_in_bed_(minutes)": "time_in_bed",
    "efficiency_(%)": "sleep_efficiency",
    "sleep_onset_latency_(minutes)": "sleep_latency",
    "deep_sleep_duration_(minutes)_(minutes)": "deep_sleep",
    "rem_duration_(minutes)": "rem_sleep",
    "awake_percentage_(%)": "awake_pct",
    "light_sleep_percentage_(%)": "light_sleep_pct"
})
def classify_efficiency(eff):
    if eff >= 90:
        return "Excellent"
    elif eff >= 80:
        return "Good"
    elif eff >= 70:
        return "Fair"
    else:
        return "Poor"

df_public["sleep_quality"] = df_public["sleep_efficiency"].apply(classify_efficiency)
df_public["sleep_eff_norm"] = df_public["sleep_efficiency"] / 100
df_public["sleep_duration_hr"] = df_public["sleep_duration"] / 60
df_public["time_in_bed_hr"] = df_public["time_in_bed"] / 60

df_public.head()
df_public.info()

```

```python
df_public.columns = df_public.columns.str.strip().str.lower().str.replace(" ", "_")

df_sleep = df_public.copy()

def bucket_sleep(h):
    if h < 5:
        return "Very Short"
    elif 5 <= h < 6.5:
        return "Short"
    elif 6.5 <= h <= 8:
        return "Normal"
    else:
        return "Long"

df_sleep["sleep_bucket"] = df_sleep["sleep_duration"].apply(bucket_sleep)
df_sleep["sleep_quality_norm"] = df_sleep["sleep_quality"] / 10
df_sleep["screen_time_norm"] = df_sleep["screen_time"] / 10
df_sleep["gender"] = df_sleep["gender"].replace("Other", "Nonbinary")
df_sleep.head()
df_sleep.info()
```
